```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import os
```

```python
# Lista de temporadas e seu padrão correspondente nas URLs
seasons = ['atual', '2022-2023', '2021-2022', '2020-2021']
url_patterns = [
    'https://fbref.com/pt/comps/11/cronograma/Serie-A-Resultados-e-Calendarios',
    'https://fbref.com/pt/comps/11/2022-2023/cronograma/2022-2023-Serie-A-Resultados-e-Calendarios',
    'https://fbref.com/pt/comps/11/2021-2022/cronograma/2021-2022-Serie-A-Resultados-e-Calendarios',
    'https://fbref.com/pt/comps/11/2020-2021/cronograma/2020-2021-Serie-A-Resultados-e-Calendarios'
]

# Caminho base para salvar os arquivos
output_base_path = '/content/drive/My Drive/Master/Games/'

# Assegurar que o diretório de saída existe
os.makedirs(output_base_path, exist_ok=True)

# Iterar sobre cada temporada e sua URL correspondente
for season, url in zip(seasons, url_patterns):
    response = requests.get(url)
    if response.status_code == 200:
        # Extrair a tabela com pandas
        list_of_dfs = pd.read_html(response.text)
        if list_of_dfs:
            # Supomos que a tabela desejada seja sempre a primeira
            current_table = list_of_dfs[0]

            file_name = f"serie_a_{season}_resultados.csv"
            output_file_path = os.path.join(output_base_path, file_name)

            # Salvar o DataFrame em um arquivo CSV
            current_table.to_csv(output_file_path, index=False)
            print(f'Arquivo salvo com sucesso em: {output_file_path}')
        else:
            print(f"Nenhuma tabela encontrada em {url}")
    else:
        print(f"Falha ao acessar {url}")
```
