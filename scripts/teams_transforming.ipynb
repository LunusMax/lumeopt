**ACESSO AOS ARQUIVOS:**

```python
# Verificar e criar a pasta de saída se não existir
if not os.path.exists(output_base_path):
    os.makedirs(output_base_path)

if not os.path.exists(output_gk):
    os.makedirs(output_gk)

# Sufixos das temporadas
season_suffixes = ['2023-2024', '2022-2023', '2021-2022', '2020-2021']

# Lista de times
team_list = [
    'atalanta', 'bologna', 'cagliari', 'cremonese', 'empoli', 'fiorentina',
    'frosinone', 'genoa', 'hellas-verona', 'internazionale', 'juventus',
    'lazio', 'lecce', 'milan', 'monza', 'napoli', 'parma', 'roma',
    'salernitana', 'sampdoria', 'sassuolo', 'spezia', 'torino', 'udinese',
    'venezia'
]
```

**PROCESSAMENTO DOS DADOS:**

```python
# Processar cada arquivo
for team in team_list:
    for season in season_suffixes:
        file_name = f"data_{team}_{season}.csv"
        file_path = os.path.join(base_path, file_name)

        # Verificar se o arquivo existe antes de tentar abri-lo
        if os.path.exists(file_path):
            data2 = pd.read_csv(file_path)

            # Remover a primeira linha do dataframe
            linha_0 = data2.iloc[0].array
            data_sem_linha_0 = data2.drop(index=[0])
            data_sem_linha_0.columns = linha_0

            # Limpeza de dados, transformações, etc.

            # Dividir o DataFrame em três partes
            df_part1 = data_sem_linha_0.iloc[:, 0:33]
            df_part3 = data_sem_linha_0.iloc[:, 63:79]

            # Combinar partes 1 e 3 e apagar a coluna [1]
            processed_data = pd.concat([df_part1, df_part3], axis=1)
            processed_data = processed_data.drop(processed_data.columns[1], axis=1)

            # Preencher espaços com 0
            processed_data.fillna(0, inplace=True)

            # Remover linhas dos totais
            processed_data = processed_data[~processed_data[processed_data.columns[0]].isin(["Total do time", "Total de adversários"])]

            # Alterar atributo idade
            processed_data['Idade'] = processed_data['Idade'].apply(lambda x: str(x).split('-')[0] if isinstance(x, str) else x)

            # Caminho do arquivo de saída
            output_file_path = os.path.join(output_base_path, file_name)

            # Salvar o DataFrame processado
            processed_data.to_csv(output_file_path, index=False)
            print(f"Arquivo processado e salvo: {output_file_path}")

        else:
            print(f"Arquivo não encontrado: {file_path}")

  #   break  # Descomentar quando for testar alteração
```

```python
# Processar tabelas de goleiros
for team in team_list:
    for season in season_suffixes:
        file_name = f"data_{team}_{season}.csv"
        file_path = os.path.join(base_path, file_name)

        # Verificar se o arquivo existe antes de tentar abri-lo
        if os.path.exists(file_path):
            data = pd.read_csv(file_path)

            # Remover a primeira linha do dataframe e usar como cabeçalho
            data.columns = data.iloc[0]
            data = data.iloc[1:].reset_index(drop=True)

            # Selecionar as colunas específicas para goleiros
            df_goleiros = data.iloc[:, 34:57]

            # Preencher espaços vazios com 0
            df_goleiros.fillna(0, inplace=True)
            df_goleiros = df_goleiros[~df_goleiros[df_goleiros.columns[0]].isin(["Total do time", "Total de adversários", 0])]

            # Alterar atributo idade
            df_goleiros['Idade'] = df_goleiros['Idade'].apply(lambda x: str(x).split('-')[0] if isinstance(x, str) else x)

            # Construir o caminho do arquivo de saída correto
            output_file_path = os.path.join(output_gk, file_name)  # Caminho correto para salvar o arquivo

            # Salvar o DataFrame processado
            df_goleiros.to_csv(output_file_path, index=False)
            print(f"Arquivo processado e salvo: {output_file_path}")

        else:
            print(f"Arquivo não encontrado: {file_path}")

# O break foi movido para comentário como você pediu, para futuras referências quando for testar
# break
```

# **Gerar um único arquivo:**

```python
# Listar todos os arquivos csv no diretório
csv_files = [file for file in os.listdir(path_players) if file.endswith('.csv')]

# Caminho a ser salvo
output_players_table = '/content/drive/My Drive/Master/Players/table_players.csv'

# Lista para armazenar cada DataFrame
dataframes_list = []

# iterar todos os arquivos csv, ler e adicionar à lista
for file in csv_files:
  file_path = os.path.join(path_players, file)
  df = pd.read_csv(file_path)
  dataframes_list.append(df)

# Concatenar todos os DataFrames na lista em um único DataFrame
final_dataframe = pd.concat(dataframes_list, ignore_index=True)

# Salvar o novo DataFrame
final_dataframe.to_csv(output_players_table, index=False)

# Exibir mensagem de confirmação e head
print(f'Arquivo salvo com sucesso em: {output_players_table}')
print(final_dataframe.head())

```

```python
# Copiar tabela para gerar a final
copy_file_path = '/content/drive/My Drive/Master/Players/final_table_players.csv'
players_data = pd.read_csv(output_players_table)

# Agrupar os dados dos jogadores
players_data = players_data.groupby('Jogador').sum()

# Copiar como saída
players_data.to_csv(copy_file_path, index=False)
print(f'Cópia criada com sucesso em: {copy_file_path}')
```
