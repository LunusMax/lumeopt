```python
# Base URL
base_url = 'https://fbref.com/pt/equipes/'

# Suffixes for the desired seasons
season_suffixes = ['', '2022-2023', '2021-2022', '2020-2021']

# List of teams
team_list = [
    '922493f3/Atalanta-Estatisticas', '1d8099f8/Bologna-Estatisticas',
    'c4260e09/Cagliari-Estatisticas', '9aad3a77/Cremonese-Estatisticas',
    'a3d88bd8/Empoli-Estatisticas', '421387cf/Fiorentina-Estatisticas',
    '6a7ad59d/Frosinone-Estatisticas', '658bf2de/Genoa-Estatisticas',
    '0e72edf2/Hellas-Verona-Estatisticas', 'd609edc0/Internazionale-Estatisticas',
    'e0652b02/Juventus-Estatisticas', '7213da33/Lazio-Estatisticas',
    'ffcbe334/Lecce-Estatisticas', 'dc56fe14/Milan-Estatisticas',
    '21680aa4/Monza-Estatisticas', 'd48ad4ff/Napoli-Estatisticas',
    'eab4234c/Parma-Estatisticas', 'cf74a709/Roma-Stats',
    'c5577084/Salernitana-Stats', '8ff9e3b3/Sampdoria-Estatisticas',
    'e2befd26/Sassuolo-Estatisticas', '68449f6d/Spezia-Stats',
    '105360fe/Torino-Stats', '04eea015/Udinese-Stats',
    'af5d5982/Venezia-Estatisticas'
]

# New folder path for saving CSV files
new_folder_path = '/content/drive/My Drive/Master/Games/'

# Iterate over the team list
for current_team in team_list:
    team_id, team_name = current_team.split('/')
    team_name = team_name.replace("-Estatisticas", "").replace("-Stats", "")

    # Iterate over the season suffixes
    for season in season_suffixes:
        complete_url = f'{base_url}{team_id}/{season}/{team_name}-Estatisticas' if season else f'{base_url}{team_id}/{team_name}-Estatisticas'

        # Make the page content request
        response = requests.get(complete_url)

        # Check if the request was successful
        if response.status_code == 200:
            tables = pd.read_html(response.text)
            if len(tables) > 1:
                second_table = tables[1]  # Second table
                # Filtering rows where the "Camp." column contains "Serie A"
                if 'Camp.' in second_table.columns:
                    filtered_table = second_table[second_table['Camp.'] == "Serie A"]
                else:
                    filtered_table = pd.DataFrame()  # Empty DataFrame if column or condition does not match

            file_name = f'{new_folder_path}data_{team_name.lower()}_{season if season else "current"}_serie_a.csv'
            filtered_table.to_csv(file_name, index=False)
            print(f'Serie A data for team {team_name} for the season {season if season else "current"} saved successfully in file {file_name}')
        else:
            print(f'Failed to access the page for team {team_name} for the season {season if season else "current"}:', response.status_code)

        # Delay between requests to avoid blocking
        time.sleep(10)
```
