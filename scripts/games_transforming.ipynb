**ACESSO AOS ARQUIVOS:**

```python
# Lista com os nomes dos arquivos
file_names = [
    'serie_a_2023-2024_resultados.csv',
    'serie_a_2022-2023_resultados.csv',
    'serie_a_2021-2022_resultados.csv',
    'serie_a_2020-2021_resultados.csv'
]
```

**PROCESSAMENTO DOS DADOS:**

```python
# Índices das colunas a serem removidas
cols_to_remove = [0, 1, 2, 3, 5, 7, 9, 10, 11, 12, 13]

# Função para converter o resultado em classe
def resultado_para_classe(resultado):
    # Verificar se o resultado é nulo ou não é uma string
    if pd.isna(resultado) or not isinstance(resultado, str):
        return None  # Retorna None ou poderia ser um código específico para resultados ausentes
    # Substituir en dash por hífen normal
    resultado = resultado.replace('–', '-')
    gols_casa, gols_visitante = map(int, resultado.split('-'))
    if gols_casa > gols_visitante:
        return 1  # Vitória do time da casa
    elif gols_casa < gols_visitante:
        return 2  # Vitória do visitante
    else:
        return 0  # Empate
```

```python
# Iterar sobre cada arquivo
for file_name in file_names:
    # Construir o caminho completo do arquivo
    file_path = os.path.join(base_path, file_name)

    # Carregar o arquivo CSV
    df = pd.read_csv(file_path)

    # Aplicar a função de conversão
    df['Classe'] = df['Resultado'].apply(resultado_para_classe)

    # Remover colunas pelo índice
    df.drop(df.columns[cols_to_remove], axis=1, inplace=True, errors='ignore')
    df.drop(df.columns[1], axis=1, inplace=True, errors='ignore')

    # Apagar linhas vazias
    df.dropna(how='all', inplace=True)

    # Caminho de saída
    output_file_path = os.path.join(output_base_path, file_name)

    # Salvar o DataFrame alterado
    df.to_csv(output_file_path, index=False)

    print(f'Alterações salvas no arquivo: {output_file_path}')
```

```python
# Tratar o arquivo diferente de forma especial (2022-2023)
# Iterar sobre cada arquivo
file_name = 'serie_a_2022-2023_resultados.csv'
base_path = '/content/drive/My Drive/Master/Games/'
file_path = os.path.join(base_path, file_name)

cols_to_remove = [0, 1, 2, 3, 5, 7, 9, 10, 11, 12, 13]
cols_to_remove_modified = [x + 1 for x in cols_to_remove]

# Carregar o arquivo CSV
df = pd.read_csv(file_path)

# Aplicar a função de conversão
df['Classe'] = df['Resultado'].apply(resultado_para_classe)

# Remover colunas pelo índice
df.drop(df.columns[cols_to_remove_modified], axis=1, inplace=True, errors='ignore')
df.drop(df.columns[2], axis=1, inplace=True, errors='ignore')
df.drop(df.columns[0], axis=1, inplace=True, errors='ignore')

# Apagar linhas vazias
df.dropna(how='all', inplace=True)

# Caminho de saída
output_file_path = os.path.join(output_base_path, file_name)

# Salvar o DataFrame alterado
df.to_csv(output_file_path, index=False)

print(f'Alterações salvas no arquivo: {output_file_path}')
```
